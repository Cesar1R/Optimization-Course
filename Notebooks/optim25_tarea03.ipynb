```python
def back_suficiente(a_init = 1.0, rho = 0.75, c1 = 0.1, xk= None, f=None, fk = None, grad_xk = None, pk = None, N=500):
    if rho <= 0 or 1 <= rho: 
        raise ValueError("El valor de rho no cumple con las especificaciones")
    
    if c1 <= 0 or 1 <= c1: 
        raise ValueError("El valor de c1 esta fuera del rango permitido") 

    if fk == None: 
        fk = f(xk) 
    
    a = a_init
    grad_xk = np.asarray(grad_xk)
    pk = np.asarray(pk) 

    for i in range(N): 
        if f(xk + a * pk) <= f(xk) + c1 * a * grad_xk.T @ pk: 
            return a, i 

        a  = rho*a 

    return a, i  


def desc_max (f, grad_f, x0, tol = eps, N = 5000, a_init = 1.0, rho = 0.75, c1 = 0.1, Ngs = 500):
    if rho <= 0 or 1 <= rho: 
        raise ValueError("El valor de rho no cumple con las especificaciones")
    
    if c1 <= 0 or 1 <= c1: 
        raise ValueError("El valor de c1 esta fuera del rango permitido") 
    
    if tol <= 0:
        raise ValueError("El valor de la tolerancia es invalido") 

    xk = np.asarray(x0)

    arr = [(x0, a_init, 0)]

    for i in range(N):
        gk = grad_f(xk)
        
        if np.linalg.norm(gk) < tol: 
            return xk, i, True, arr
        
        pk = -gk 

        a, k = back_suficiente(a_init=a_init, rho=rho, c1=c1, xk=xk, f=f, fk=f(xk), grad_xk=gk, pk=pk)

        xk = xk + a*pk

        if len(x0) == 2: 
            arr.append((xk, a, k))

    return xk, N, False, arr 


```

```python
def gen_descenso_coordenadas(f, x0, N=40000, tol=1e-8, xl=-1., xu=1., Ngs=450):
    xk = np.asarray(x0)
    n = len(xk)
    
    indices = np.arange(n)

    vec_i = np.zeros(n)

    arr = [x0] 
    for i in range(N):
        x = xk.copy()  
        np.random.shuffle(indices) 

        for j in indices:
            vec_i[j] = 1
            f_aux = lambda t: f(xk + t*vec_i)
            
            aux = met_seccion_dorada(f_aux, xl, xu, tol, Ngs)
            xk[j] += aux[0]

            vec_i[j] = 0
        
        x_copia = xk.copy()
        if n == 2: 
            arr.append(x_copia)  

        if np.linalg.norm(xk - x) < tol:
            return xk, i, True, arr

    return xk, N, False, arr
```

## Ejercicio 3 (1.5 puntos)

Sea $f(\mathbf{x}) = f(x_1, x_2) = 5 + x_1^2 + x_2^2$. Si $\mathbf{x}_0 = (-1,1)^\top$,
$\mathbf{p}_0 = (1,0)$ y $c_1=10^{-4}$.
Verifique de $\mathbf{p}_0$ es una dirección de descenso y encuentre el valor 
más grande $\alpha>0$ que satisface la condición de descenso suficiente:

$$ f(\mathbf{x}_0 + \alpha\mathbf{p}_0) \leq 
f(\mathbf{x}_0) + c_1 \alpha\mathbf{p}_0^\top\nabla f(\mathbf{x}_0). $$ 

### Solución:

Para verificar que $p0$ es una dirección de descenso se debe cumplir que 

$$ \nabla f(x_0)^T p_0 < 0. $$

Comencemos Calculando el gradiente. 

$$
\partial_{x_1} f (x_1, x_2) = 2x_1.  
$$

Además, 

$$
\partial_{x_2} f(x_1, x_2) = 2x_2, 
$$

por lo tanto, 

$$
\nabla f(x) = (2x_1, 2x_2), 
$$

luego, 

$$
\nabla f(x_0)^T p_0 = (-2, 2) \cdot (1, 0) = -2,
$$

en particular se cumple que $\nabla f(x_0)^T p_0 < 0$. Con lo que conlcuimos la primera parte, afirmando que $p_0$ es una dirección de descenso.  

Por otro lado,  comencemos evaluando el lado izquierdo de la desigualdad

$$
f(x_0 + \alpha p_0) = 5 + (1 - 2 \alpha + \alpha^2) + 1  = 7 - 2 \alpha + \alpha^2, 
$$

siguiendo con el lado derecho obtenemos que

$$
 f(x_0) + c_1 \alpha p_0^T = f(-1, 1) + 10^{-4}\alpha(-2) = 7  - (2 \cdot 10^{-4}) \alpha
$$

Sustituyendo en la desigualdad de interés obtenemos que  
$$
\begin{align*}
f(x_0 + \alpha p_0) \leq  f(x_0) + c_1 \alpha p_0^T \nabla f(x_0) 
&\Leftrightarrow  2\alpha + 14 \cdot 10^{-4} \alpha + \alpha^2 \leq 0 \\ 
&\Leftrightarrow  (2 \cdot 10^{-4} - 2)\alpha + \alpha^2 \leq 0 \\ 
&\Leftrightarrow \alpha( \alpha + (2 \cdot 10^{-4} - 2)) \leq 0\\
\end{align*}
$$


Notemos que si $\alpha < 0$ entonces toda la expresión se vuelve positiva, pues  el producto de dos n\'umeros negativos es positivo y tenemos que $\alpha (2 \cdot 10^{-4} - 2)  > 0 $ al igual que $\alpha^2$. Por otro lado si $\alpha >  (2 - 2 \cdot 10^{-4} )$ esto implica que  que $\alpha - (2 \cdot 10^{-4} - 2) = \alpha + 2 - 2\cdot 10^{-4} > 0 $, por lo tanto, todo el producto es positivo. 

De lo anterior obtenemos que la desigualdad se cumple para $0 \leq \alpha \leq (2 - 2 \cdot 10^{-4} )$, concluyendo que el valor más grande que satisface la condición de descenso es $2- 2 \cdot 10^{-4} $


## Ejercicio 4 (1.5 puntos)

Sea $f:\mathbb{R}^n \rightarrow \mathbb{R}$ y $\mathbf{S}$ una matriz no singular de tamaño $n \times n$. 
Si $\mathbf{x} = \mathbf{S}\mathbf{y}$ para $\mathbf{y} \in \mathbb{R}^n$ y definimos
$g(\mathbf{y}) = f(\mathbf{S}\mathbf{y})$, aplicando la regla de la cadena muestre que 

$$ \nabla g(\mathbf{y}) =  \mathbf{S}^\top \nabla f(\mathbf{x}). $$

Entonces aplicando el método de máximo descenso a la función $g$ es 

$$ \mathbf{y}_{k+1} = \mathbf{y}_k - \alpha_k \mathbf{S}^\top \nabla f(\mathbf{S}\mathbf{y}_k).$$

Multiplicando por $\mathbf{S}$ ambos miembros de la ecuación y usando la notación
$\mathbf{x}_k =\mathbf{S} \mathbf{y}_k$:

$$ \mathbf{x}_{k+1} =  \mathbf{S} \mathbf{y}_{k+1} = 
\mathbf{S} \mathbf{y}_k - \alpha_k \mathbf{S}\mathbf{S}^\top \nabla f(\mathbf{S}\mathbf{y}_k)
= \mathbf{x}_k - \alpha_k \mathbf{S}\mathbf{S}^\top \nabla f(\mathbf{x}_k).$$

Si $\mathbf{D} = \mathbf{S}\mathbf{S}^\top$, obtenemos el método de *gradiente escalado*:

$$ \mathbf{x}_{k+1} = \mathbf{x}_k - \alpha_k \mathbf{D} \nabla f(\mathbf{x}_k).$$

Muestre que $-\mathbf{D} \nabla f(\mathbf{x}_k)$ es una dirección de descenso.

### Solución:

Sean $f$ y $\mathbf{S}$ como en el enunciado. Por la regla de la cadena tenemos que

$$
D g(y) = D f(x) \circ D x = D f(x) \circ D Sy = \nabla f(x)^T S 
$$

obteniendo asi que 

$$
\nabla g(y) = \left ( \nabla f(x)^T S \right)^T  = S^T \nabla f(x),
$$

concluyendo así la primera parte. 

Por otro lado, obvervemos que 

$$
\nabla f(x_k)^T (-D\nabla f(x_k)) = - \nabla f(x_k)^T D\nabla f(x_k)  = -\nabla f(x_k)^T SS^T\nabla f(x_k)
$$

Notemos que si podemos demostrar que $SS^T$ es positiva definida habremos terminado la demostración, pues eso implicaria que la exprersión anterior es negativa. 

Por hipótesis sabemos que $S$ es no singular, entonces la matriz $SS^T$ es una matriz definida positiva. Concluyendo así que $\nabla f(x_k)^T (-D\nabla f(x_k)) < 0$, por lo tanto $-D\nabla f(x_k)$ es una dirección de descenso.
